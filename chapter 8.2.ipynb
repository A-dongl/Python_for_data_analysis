# 8.2 데이터 합치기

- 8.2.1 데이터베이스 스타일로 DataFrame합치기

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                   'data1': range(7)})

df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                   'data2': range(3)})

df1
```

```python
df2
```

```python
pd.merge(df1, df2)  # 중복된 column을 키로 사용한다.
```

```python
pd.merge(df1, df2, on='key') # 합쳐진 기준 column 을 명시
```

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a','c', 'a', 'a', 'b'],
                   'data1': range(7)})

df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],
                   'data2': range(3)})

pd.merge(df3, df4, left_on = 'lkey', right_on = 'rkey')
```

```python
pd.merge(df1, df2, how = 'outer')  # 양쪽 테이블에 존재하는 모든 키 조합을 사용한다
```

```python
df1 = pd.DataFrame({'key': ['b','b', 'a', 'c', 'a', 'b'],
                   'data1': range(6)})

df2 = pd.DataFrame({'key': ['a', 'b', 'a', 'b', 'd'],
                   'data2': range(5)})

df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='key', how='left')
```

```python
pd.merge(df1, df2, how='inner')
```

```python
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar'],
                    'key2': ['one', 'two', 'one'],
                    'lval': [1,2,3]})

right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                     'key2': ['one', 'one', 'one', 'two'],
                     'rval': [4, 5, 6, 7]})

pd.merge(left, right, on=['key1', 'key2'], how='outer')
```

```python
pd.merge(left, right, on='key1')
```

```python
pd.merge(left, right, on='key1', suffixes=('_left', '_right'))  #suffixes 를 이용해서 컬럼 이름 뒤에 붙일 문자열 설정 가능
```

- 8.2.2 색인 병합하기

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],
                     'value': range(6)})

right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])

left1
```

```python
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)  # 교집합 
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')  # 합집합
```

```python
lefth = pd.DataFrame({'key1': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada'],
                     'key2': [2000, 2001, 2002, 2001, 2002],
                     'data':np.arange(5.)})

righth = pd.DataFrame(np.arange(12).reshape((6,2)),
                     index = [['Nevada', 'Nevada', 'Ohio', 'Ohio', 'Ohio', 'Ohio'],
                             [2001, 2000, 2000, 2000, 2001, 2002]],
                     columns=['event1', 'event2'])

lefth
```

```python
righth
```

```python
pd.merge(lefth, righth, left_on=['key1', 'key2'],
        right_index=True, how='outer')
```

```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                    index=['a', 'c', 'e'],
                    columns=['Ohio', 'Nevada'])

left2
```

```python
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                     index=['b', 'c', 'd', 'e'],
                     columns=['Missouri', 'Alabama'])

right2
```

```python
pd.merge(left2, right2, how='outer', left_index=True, right_index=True)
```

```python
left2.join(right2, how='outer')
```

```python
left1.join(right1, on='key')
```

```python
another = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [16., 17.]],
                      index=['a', 'c', 'e', 'f'],
                      columns=['New York', 'Oregon'])

another
```

```python
left2.join([right2, another])
```

```python
left2.join([right2, another], how='outer')
```

- 8.2.3 축 따라 이어붙이기

```python
arr = np.arange(12).reshape((3, 4))
arr
```

```python
np.concatenate([arr, arr], axis=1)
```

```python
s1 = pd.Series([0, 1], index=['a', 'b'])
s2 = pd.Series([2, 3, 4], index=['c', 'd', 'e'])
s3 = pd.Series([5, 6], index=['f', 'g'])
```

```python
pd.concat([s1, s2, s3])  # concat 함수에 전달하면 값과 색인을 연결
```

```python
pd.concat([s1, s2, s3], axis=1)
```

```python
s4 = pd.concat([s1, s3])
```

```python
s4
```

```python
pd.concat([s1, s4], axis=1)
```

```python
pd.concat([s1, s4], axis=1, join='inner')
```

```python
pd.concat([s1, s4], axis=1).reindex(['a', 'c', 'b', 'e'])
# join_axes 는 이제 사용할 수 없고 reindex를 사용하여 순서를 정할 수 있다
```

```python
result = pd.concat([s1, s1, s3], keys=['one', 'two', 'three'])
```

```python
result
```

```python
result.unstack()
```

```python
pd.concat([s1, s2, s3], axis=1, keys=['one', 'two', 'three']) #keys 는 컬럼의 이름
```

```python
df5 = pd.DataFrame(np.arange(6).reshape(3,2), index=['a', 'b', 'c'],
                  columns=['one', 'two'])

df6 = pd.DataFrame(5+np.arange(4).reshape(2,2), index=['a', 'c'],
                  columns=['three', 'four'])

df5
```

```python
df6
```

```python
pd.concat([df5, df6], axis=1, keys=['level1', 'level2'])
```

```python
pd.concat({'level1': df5, 'level2': df6}, axis=1)  # dict 형태로 받을 수도 있다
```

```python
pd.concat([df5, df6], keys=['level1', 'level2'],
         names=['upper', 'lower'])
```

```python
df7 = pd.DataFrame(np.random.randn(3, 4), columns=['a', 'b', 'c', 'd'])

df8 = pd.DataFrame(np.random.randn(2, 3), columns=['b', 'd', 'e'])

df7
```

```python
df8
```

```python
pd.concat([df7, df8], ignore_index=True)  # 합집합으로 가져오기
```

- 8.2.4 겹치는 데이터 합치기

```python
a = pd.Series([np.nan, 2.5, np.nan, 3.5, 4.5, np.nan],
             index=['f', 'e', 'd', 'c', 'b', 'a'])

b = pd.Series(np.arange(len(a), dtype=np.float64),
             index=['f','e', 'd', 'c', 'b', 'a'])

b[-1] = np.nan

a
```

```python
b
```

```python
np.where(pd.isnull(a), b, a)
```

```python
b[:-2].combine_first(a[2:])   #data 정렬가능
```

```python
df9 = pd.DataFrame({'a': [1., np.nan, 5., np.nan],
                  'b': [np.nan, 2., np.nan, 6.],
                  'c': range(2, 18, 4)})

df10 = pd.DataFrame({'a': [5., 4., np.nan, 3., 7.],
                    'b': [np.nan, 3., 4., 6., 8.]})

df9
```

```python
df10
```

```python
df9.combine_first(df10)  # 채워넣기!!
```
